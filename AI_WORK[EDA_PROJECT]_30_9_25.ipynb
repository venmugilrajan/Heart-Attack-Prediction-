```python
df=pd.read_csv("heart_attack.csv")

df[['Systolic', 'Diastolic']]=df['Blood Pressure'].str.split('/',expand=True)
df['Systolic'] = pd.to_numeric(df['Systolic'])
df['Diastolic'] = pd.to_numeric(df['Diastolic'])
df = df.drop(columns=['Blood Pressure'])
```

```python
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})
df
```

```python
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].median())
df
```

```python
df['Pulse Pressure'] = df['Systolic'] - df['Diastolic']
df['Cholesterol_BMI_Ratio'] = df['Cholesterol'] / (df['BMI'] + 1)
df
```

```python
plt.hist(df['BMI'], bins=10, color='lightgreen', edgecolor='black')
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
risk_counts = df['Heart Attack Risk'].value_counts()
plt.bar(risk_counts.index, risk_counts.values, color=['green','red'])
plt.xticks([0, 1], ['No Risk', 'Risk'])
plt.title("Heart Attack Risk Count")
plt.ylabel("Number of Patients")
plt.show()
```

```python
if 'Systolic' in df.columns and 'Diastolic' in df.columns:
    df['Pulse_Pressure'] = df['Systolic'] - df['Diastolic']

if 'Age' in df.columns:
    df['Age_Group'] = pd.cut(df['Age'], bins=[0,30,45,60,120],
                             labels=['0-30','31-45','46-60','61+'])

risk_cols = ['Diabetes','Family History','Smoking','Obesity']
df['Risk_Sum'] = df[risk_cols].sum(axis=1)
df
```

```python
features = ['Age','Cholesterol','BMI','Risk_Sum']
target = 'Heart Attack Risk'

plt.figure(figsize=(12,8))
plotnumber = 1
for col in features:
    plt.subplot(2, 2, plotnumber)
    sns.kdeplot(df[df[target]==0][col], label='No Risk', fill=True, alpha=0.5)
    sns.kdeplot(df[df[target]==1][col], label='High Risk', fill=True, alpha=0.5)
    plt.title(f"Distribution of {col} by Heart Attack Risk")
    plt.xlabel(col)
    plt.legend()
    plotnumber += 1
plt.tight_layout()
plt.show()
```
