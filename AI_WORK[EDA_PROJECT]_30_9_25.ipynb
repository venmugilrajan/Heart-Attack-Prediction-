```python
df=pd.read_csv("/content/heart_attack.csv")

df[['Systolic', 'Diastolic']]=df['Blood Pressure'].str.split('/',expand=True)
df['Systolic'] = pd.to_numeric(df['Systolic'])
df['Diastolic'] = pd.to_numeric(df['Diastolic'])
df = df.drop(columns=['Blood Pressure'])
```

```python
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})
df
```

```python
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].median())
df
```

```python
df['Pulse Pressure'] = df['Systolic'] - df['Diastolic']
df['Cholesterol_BMI_Ratio'] = df['Cholesterol'] / (df['BMI'] + 1)
df
```

```python
plt.hist(df['BMI'], bins=10, color='lightgreen', edgecolor='black')
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
risk_counts = df['Heart Attack Risk'].value_counts()
plt.bar(risk_counts.index, risk_counts.values, color=['green','red'])
plt.xticks([0, 1], ['No Risk', 'Risk'])
plt.title("Heart Attack Risk Count")
plt.ylabel("Number of Patients")
plt.show()
```

```python
avg_chol_by_age = df.groupby('Age')['Cholesterol'].mean()
plt.plot(avg_chol_by_age.index, avg_chol_by_age.values, marker='x', color='green',markeredgecolor='blue',markersize=8)
plt.title("Average Cholesterol by Age")
plt.xlabel("Age")
plt.ylabel("Average Cholesterol")
plt.grid(True)
plt.show()
```
